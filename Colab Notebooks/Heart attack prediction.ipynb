```python
Scaler = StandardScaler()
```

```python
y = df["output"]
df.drop("output",axis=1,inplace=True)
```

```python
cont_cols.remove("output")
```

```python
cont_cols
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
RF = RandomForestClassifier()
```

```python
scaled = Scaler.fit_transform(df[cont_cols])
```

```python
scaled_df = pd.DataFrame(scaled,columns=cont_cols)
```

```python
scaled_df = pd.concat([scaled_df,df[cat_cols]],axis=1,sort=False)
```

```python
scaled_df.describe().T
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(scaled_df,y,test_size=0.3)
```

```python
RF.fit(X_train,y_train)
```

```python
base_model = RF.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score as acc
```

```python
acc(y_test,base_model)
```

```python
def model(X_train,X_test, model = RandomForestClassifier(),y_train=y_train, y_test=y_test):
  RF = model
  RF.fit(X_train,y_train)
  y_pred = RF.predict(X_test)
  error = mae(y_test,y_pred)
  accuracy = acc(y_test,y_pred)
  f1 = f1_score
  print(f"error : {error}")
  print(f"accuracy = {accuracy}")
```

```python
model(X_train,X_test)
```
