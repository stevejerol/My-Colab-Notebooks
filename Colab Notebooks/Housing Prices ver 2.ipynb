```python
ntrain = df_train.shape[0]
ntest = df_test.shape[0]
y_train = df_train.SalePrice.values
all_data = pd.concat((df_train, df_test)).reset_index(drop=True)
all_data.drop(['SalePrice'], axis=1, inplace=True)
```

```python
all_data.shape
```

```python
total = df_train.isnull().sum().sort_values(ascending=False)
percent = (df_train.isnull().sum()/df_train.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data.head(20)
```

```python
all_data["PoolQC"] = all_data["PoolQC"].fillna("None")
```

```python
all_data["MiscFeature"] = all_data["MiscFeature"].fillna("None")
all_data["Alley"] = all_data["Alley"].fillna("None")
all_data["Fence"] = all_data["Fence"].fillna("None")
all_data["FireplaceQu"] = all_data["FireplaceQu"].fillna("None")
```

```python
all_data["LotFrontage"].fillna(all_data["LotFrontage"].mean())
```

```python
for cols in ('GarageType', 'GarageFinish', 'GarageQual', 'GarageCond'):
  all_data[cols]=all_data[cols].fillna("None")
```

```python
for cols in ('GarageYrBlt', 'GarageArea', 'GarageCars'):
  all_data[cols]=all_data[cols].fillna(0)
```

```python
for col in ('BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF','TotalBsmtSF', 'BsmtFullBath', 'BsmtHalfBath'):
    all_data[col] = all_data[col].fillna(0)
```

```python
for col in ('BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2'):
    all_data[col] = all_data[col].fillna('None')
```

```python
all_data["MasVnrType"] = all_data["MasVnrType"].fillna("None")
all_data["MasVnrArea"] = all_data["MasVnrArea"].fillna(0)
```

```python
total = all_data.isnull().sum().sort_values(ascending=False)
percent = (all_data.isnull().sum()/all_data.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data.head(10)
```

```python
all_data["LotFrontage"] = all_data["LotFrontage"].fillna(all_data["LotFrontage"].mean())
```

```python
all_data['SaleType'].dropna()
```

```python
all_data['MSZoning'] = all_data['MSZoning'].fillna(all_data['MSZoning'].mode()[0])
```

```python
all_data = all_data.drop(['Utilities'], axis=1)
```

```python
all_data["Functional"] = all_data["Functional"].fillna(all_data["Functional"].mode()[0])
```

```python
for cols in ("Electrical","KitchenQual","Exterior2nd","Exterior1st","SaleType"):
  all_data[cols] = all_data[cols].fillna(all_data[cols].mode()[0])
```

```python
all_data.isna().any().sum()
```

```python
all_data['MSSubClass'] = all_data['MSSubClass'].apply(str)

all_data['OverallCond'] = all_data['OverallCond'].astype(str)

all_data['YrSold'] = all_data['YrSold'].astype(str)
all_data['MoSold'] = all_data['MoSold'].astype(str)
```

```python
from sklearn.preprocessing import LabelEncoder

s= (all_data.dtypes=="object")
object_cols = list(s[s].index)
object_cols
```

```python
for cols in object_cols:
  le = LabelEncoder()
  all_data[cols]=le.fit_transform(list(all_data[cols].values))
```

```python
from sklearn.preprocessing import StandardScaler

Scaler = StandardScaler()
Scaler.fit_transform(all_data)
```

```python
train = all_data[:ntrain]
test = all_data[ntrain:]
```

```python

from sklearn.model_selection import KFold, cross_val_score
n_folds = 5

def rmsle_cv(model):
    kf = KFold(n_folds, shuffle=True, random_state=42).get_n_splits(train.values)
    rmse= np.sqrt(-cross_val_score(model, train.values, y_train, scoring="neg_mean_squared_error", cv = kf))
    return(rmse)
```

```python
def report_best_scores(results, n_top=3):
    for i in range(1, n_top + 1):
        candidates = np.flatnonzero(results['rank_test_score'] == i)
        for candidate in candidates:
            print("Model with rank: {0}".format(i))
            print("Mean validation score: {0:.3f} (std: {1:.3f})".format(
                  results['mean_test_score'][candidate],
                  results['std_test_score'][candidate]))
            print("Parameters: {0}".format(results['params'][candidate]))
            print("")
```
