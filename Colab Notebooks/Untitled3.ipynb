```python
def brack_val(s):
  stack=""
  for i in range(len(s)):
    if stack=="":
      stack.append(s[i])
      pass
    if s[i] =="]":
      if stack[-1]=="[":
        stack.pop()
      else:
        stack.append(s[i])
    elif s[i] =="}":
      if stack[-1]=="{":
        stack.pop()
      else:
        stack.append(s[i])
    elif s[i] ==")":
      if stack[-1]=="(":
        stack.pop()
      else:
        stack.append(s[i])  
  if stack.isempty():
    print("Validated")
  else:
    print("Not Validated")
```

```python
brack_val(right)
```
